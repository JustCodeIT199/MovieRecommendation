```python
import pandas as pd
path = 'C:\\Users\\omsaw\\Project\\DataSet'
movies_metadata = pd.read_csv(path + '/movies_metadata.csv')
movies_metadata.head()
```

```python
credits = pd.read_csv(path + '/credits.csv')
credits.head()
```

```python
links = pd.read_csv(path + '/links.csv')
links.head()
```

```python
keywords = pd.read_csv(path + '/keywords.csv')
keywords.head()
```

```python
import numpy as np

# Convert 'id' columns to numeric, coercing errors to NaN
movies_metadata['id'] = pd.to_numeric(movies_metadata['id'], errors='coerce')
credits['id'] = pd.to_numeric(credits['id'], errors='coerce')
links['tmdbId'] = pd.to_numeric(links['tmdbId'], errors='coerce')

# Drop rows where 'id' or 'tmdbId' became NaN after conversion, as these cannot be merged
movies_metadata.dropna(subset=['id'], inplace=True)
credits.dropna(subset=['id'], inplace=True)
links.dropna(subset=['tmdbId'], inplace=True)

# Ensure 'id' columns are integers after handling NaNs for cleaner merging
movies_metadata['id'] = movies_metadata['id'].astype(int)
credits['id'] = credits['id'].astype(int)
links['tmdbId'] = links['tmdbId'].astype(int)

print('Data types for merge keys after cleaning:')
print(f"movies_metadata['id'] dtype: {movies_metadata['id'].dtype}")
print(f"credits['id'] dtype: {credits['id'].dtype}")
print(f"links['tmdbId'] dtype: {links['tmdbId'].dtype}")
```

Now that the ID columns are cleaned and typed correctly, I will merge `movies_metadata` with `credits`, and then the resulting dataframe with `links`.

```python
# Merge movies_metadata and credits
merged_df = pd.merge(movies_metadata, credits, on='id', how='inner')

# Merge the result with links using 'id' from merged_df and 'tmdbId' from links
merged_df = pd.merge(merged_df, links, left_on='id', right_on='tmdbId', how='inner')

# Merge the result with keywords
merged_df = pd.merge(merged_df, keywords, on='id', how='inner')

# Drop redundant ID columns from the links table
merged_df.drop(columns=['tmdbId', 'movieId'], inplace=True)

print('Shape of the merged dataframe:', merged_df.shape)
display(merged_df.head())
```

```python
columns_to_keep = ['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew', 'tagline']
movies = merged_df[columns_to_keep].copy()

print('Shape of the filtered dataframe:', movies.shape)
display(movies.head())
```

```python
import ast

#function to covert
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  L = []
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].astype(str).apply(lambda x:x.split())
```

```python
movies.isnull().sum()
```

```python
movies = movies.drop('tagline', axis=1)
```

```python
movies.isnull().sum()
```

```python
movies = movies.dropna(subset=['overview'])
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
df = movies[['id','title','tags']]
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
df.head()
```

```python
df['tags'][0]
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
df.head()
```

```python
df.shape
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2), stop_words='english')
vectorized_data = tfidf.fit_transform(df['tags'])

print("Shape of vectorized data:", vectorized_data.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectorized_data)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = df[df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list:
    print(df.iloc[i[0]].title)
```
